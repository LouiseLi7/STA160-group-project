```python
import pandas as pd
import numpy as np
h_d = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(h_d[h_d['HeartDiseaseorAttack']==0]['BMI'], color = 'green', edgecolor = 'black')
plt.hist(h_d[h_d['HeartDiseaseorAttack']==1]['BMI'], color = 'blue', edgecolor = 'black')
```

```python
BMI_h_d_quantile = pd.DataFrame({'Data':['no heart disease BMI','heart disease BMI','all BMI distribution'],
                                '25%':[np.quantile(h_d[h_d['HeartDiseaseorAttack']==0]['BMI'],0.25),
                                      np.quantile(h_d[h_d['HeartDiseaseorAttack']==1]['BMI'],0.25),
                                      np.quantile(h_d['BMI'],0.25)],
                                '50%':[np.quantile(h_d[h_d['HeartDiseaseorAttack']==0]['BMI'],0.5),
                                      np.quantile(h_d[h_d['HeartDiseaseorAttack']==1]['BMI'],0.5),
                                      np.quantile(h_d['BMI'],0.5)],
                                '75%':[np.quantile(h_d[h_d['HeartDiseaseorAttack']==0]['BMI'],0.75),
                                      np.quantile(h_d[h_d['HeartDiseaseorAttack']==1]['BMI'],0.75),
                                      np.quantile(h_d['BMI'],0.75)]})
```

```python
highBP_ratio = []
highChol_ratio = []
Smoker_ratio = []
Stroke_ratio = []
for i in range(0,2):
    highBP_ratio.append(sum(h_d[h_d['HeartDiseaseorAttack']==1]['HighBP']==1)/sum(h_d['HighBP']==i))
    highChol_ratio.append(sum(h_d[h_d['HeartDiseaseorAttack']==1]['HighChol']==1)/sum(h_d['HighChol']==i))
    Smoker_ratio.append(sum(h_d[h_d['HeartDiseaseorAttack']==1]['Smoker']==1)/sum(h_d['Smoker']==i))
    Stroke_ratio.append(sum(h_d[h_d['HeartDiseaseorAttack']==1]['Stroke']==1)/sum(h_d['Stroke']==i))
BP_Chol_Smoker_Stroke_ratios = pd.DataFrame({'Data':['heart disease HighBP','heart disease HighChol',
                                                     'heart disease Smoker','heart disease Stroke'],
                                'No indicator':[highBP_ratio[0],highChol_ratio[0],Smoker_ratio[0],Stroke_ratio[0]],
                                'Yes':[highBP_ratio[1],highChol_ratio[1],Smoker_ratio[1],Stroke_ratio[1]]})
BP_Chol_Smoker_Stroke_ratios
```
