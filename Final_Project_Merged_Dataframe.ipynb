```python
batting_all = pd.concat([Arizona_Diamondbacks, Atlanta_Braves, Baltimore_Orioles,Boston_Red_Sox,Chicago_white_Sox,Chicago_Cubs,Cincinnati_Reds,Cleaveland_Guardians,Colorado_Rockies, Detroit_Tigers,Houston_Astros,Kansas_City_Royals,Los_Angeles_Angels,Los_Angeles_Dodgers,Miami_Marlins,Milwaukee_Brewers,Minnesota_Twins,New_York_Mets,New_York_Yankees,Oakland_Athletics, Philadelphia_Phillies,Pittsburgh_Pirates,San_Diego_Padres,San_Francisco_Giants,Seattle_Mariners,StLouis_Cardinals,Tempa_Bay_Rays,Texas_Rangers,Toronto_Blue_Jays,Washington_Nationals])
batting_all = batting_all.reset_index()
TeamName = ['Arizona_Diamondbacks','Atlanta_Braves','Baltimore_Orioles','Boston_Red_Sox','Chicago_White_Sox','Chicago_Cubs','Cincinnati_Reds','Cleaveland_Guardians','Colorado_Rockies', 'Detroit_Tigers','Houston_Astros','Kansas_City_Royals','Los_Angeles_Angels','Los_Angeles_Dodgers','Miami_Marlins','Milwaukee_Brewers','Minnesota_Twins','New_York_Mets','New_York_Yankees','Oakland_Athletics', 'Philadelphia_Phillies','Pittsburgh_Pirates','San_Diego_Padres','San_Francisco_Giants','Seattle_Mariners','St.Louis_Cardinals','Tempa_Bay_Rays','Texas_Rangers','Toronto_Blue_Jays','Washington_Nationals']
for i in range(0,len(TeamName)):
    TeamName[i] = TeamName[i].replace('_',' ')
TeamMeanWar = [batting_all[batting_all['Team']==i]['WAR'].mean() for i in TeamName]
```

```python
batting = pd.concat([Arizona_Diamondbacks, Atlanta_Braves, Baltimore_Orioles,Boston_Red_Sox,Chicago_white_Sox,Chicago_Cubs,Cincinnati_Reds,Cleaveland_Guardians,Colorado_Rockies, Detroit_Tigers,Houston_Astros,Kansas_City_Royals,Los_Angeles_Angels,Los_Angeles_Dodgers,Miami_Marlins,Milwaukee_Brewers,Minnesota_Twins,New_York_Mets,New_York_Yankees,Oakland_Athletics, Philadelphia_Phillies,Pittsburgh_Pirates,San_Diego_Padres,San_Francisco_Giants,Seattle_Mariners,StLouis_Cardinals,Tempa_Bay_Rays,Texas_Rangers,Toronto_Blue_Jays,Washington_Nationals], axis=1, join="inner")
```

```python
TMW_table = pd.DataFrame({'Team':TeamName,
              'TeamMeanWar':TeamMeanWar})
```

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(range(0,len(TMW_table)),TMW_table['TeamMeanWar'], s=5)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
HCward=AgglomerativeClustering(n_clusters=3, linkage='ward')
HCward.fit(np.asarray(TMW_table['TeamMeanWar']).reshape(-1,1))
plt.scatter(range(0,len(TMW_table)),TMW_table['TeamMeanWar'], c=HCward.labels_, cmap='viridis')
plt.title('HC ward-linkage with 3 clusters')
plt.show()
```

```python
np.asarray(TMW_table['TeamMeanWar'])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(np.reshape(np.asarray(TMW_table['TeamMeanWar']), (len(TMW_table['TeamMeanWar']), 1)))

# truncated and count each cluster
plt.figure(figsize=(7, 5))
dendrogram(linked,
           p=4, 
           truncate_mode='lastp',
           color_threshold=-1,
           orientation='top',
           show_leaf_counts=True
           )
plt.show()
```
