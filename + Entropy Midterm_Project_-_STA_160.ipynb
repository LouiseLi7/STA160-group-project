```python
chol_hd = pd.crosstab(hd_data['HeartDiseaseorAttack'], hd_data['HighChol'])
chol_hd
```

```python
# entropy:
s0,s1 = chol_hd.sum(axis=1)
s2,s3 = chol_hd.sum(axis=0)
# Entropy of getting heart disease
E_hd = -s0/len(hd_data)*math.log2(s0/len(hd_data))-s1/len(hd_data)*math.log2(s1/len(hd_data))
# Entropy for high chol:
E_HC = -chol_hd[1][0]/s3*math.log2(chol_hd[1][0]/s3)-chol_hd[1][1]/s3*math.log2(chol_hd[1][1]/s3)
# Entropy for normal chol:
E_NC = -chol_hd[0][0]/s2*math.log2(chol_hd[0][0]/s2)-chol_hd[0][1]/s3*math.log2(chol_hd[0][1]/s2)
# E(heart disease|chol):
E = s3/len(hd_data)*E_HC + s2/len(hd_data)*E_NC
print(E_hd,E_HC,E_NC,E)
```

```python
BP_hd = pd.crosstab(hd_data['HeartDiseaseorAttack'], hd_data['HighBP'])
BP_hd
```

```python
# entropy:
s0,s1 = BP_hd.sum(axis=1)
s2,s3 = BP_hd.sum(axis=0)
# Entropy of getting heart disease
E_hd = -s0/len(hd_data)*math.log2(s0/len(hd_data))-s1/len(hd_data)*math.log2(s1/len(hd_data))
# Entropy for high BP:
E_HBP = -BP_hd[1][0]/s3*math.log2(BP_hd[1][0]/s3)-BP_hd[1][1]/s3*math.log2(BP_hd[1][1]/s3)
# Entropy for normal BP:
E_NBP = -BP_hd[0][0]/s2*math.log2(BP_hd[0][0]/s2)-BP_hd[0][1]/s3*math.log2(BP_hd[0][1]/s2)
# E(heart disease|BP):
E = s3/len(hd_data)*E_HBP + s2/len(hd_data)*E_NBP
print(E_hd,E_HBP,E_NBP,E)
```

```python
Smoke = pd.crosstab(hd_data['HeartDiseaseorAttack'], hd_data['Smoker'])
Smoke
```

```python
# entropy:
s0,s1 = Smoke.sum(axis=1)
s2,s3 = Smoke.sum(axis=0)
# Entropy of getting heart disease
E_hd = -s0/len(hd_data)*math.log2(s0/len(hd_data))-s1/len(hd_data)*math.log2(s1/len(hd_data))
# Entropy for high smoker:
E_S = -Smoke[1][0]/s3*math.log2(Smoke[1][0]/s3)-Smoke[1][1]/s3*math.log2(Smoke[1][1]/s3)
# Entropy for not smoker:
E_NS = -BP_hd[0][0]/s2*math.log2(Smoke[0][0]/s2)-Smoke[0][1]/s3*math.log2(Smoke[0][1]/s2)
# E(heart disease|smoker):
E = s3/len(hd_data)*E_S + s2/len(hd_data)*E_NS
print(E_hd,E_S,E_NS,E)
```

```python
Walk = pd.crosstab(hd_data['HeartDiseaseorAttack'], hd_data['DiffWalk'])
Walk
```

```python
# entropy:
s0,s1 = Walk.sum(axis=1)
s2,s3 = Walk.sum(axis=0)
# Entropy for walk:
E_W = -Walk[1][0]/s3*math.log2(Walk[1][0]/s3)-Walk[1][1]/s3*math.log2(Walk[1][1]/s3)
# Entropy for not walk:
E_NW = -Walk[0][0]/s2*math.log2(Walk[0][0]/s2)-Walk[0][1]/s3*math.log2(Walk[0][1]/s2)
# E(heart disease|walk):
E = s3/len(hd_data)*E_W + s2/len(hd_data)*E_NW
print(E_hd,E_W,E_NW,E)
```
