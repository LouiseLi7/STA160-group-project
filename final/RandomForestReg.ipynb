### Random Forest

```python
df = pd.get_dummies(batting, columns = ['Name', 'Team'])
df
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error

train, test = train_test_split(df, test_size=0.2, random_state=21)

X_train, y_train = train.copy().drop(columns=['AVG']), train['AVG']
X_test, y_test = test.copy().drop(columns=['AVG']), test['AVG']

forest = RandomForestRegressor(300, random_state=42)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Testing MSE:', mse)
acc = forest.score(X_test, y_test)
print('Accuracy:', acc)
```

```python
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(15, 10))
plot_tree(forest.estimators_[0], 
          feature_names=X_train.columns,
          class_names=['AVG'], 
          filled=True, impurity=True, 
          rounded=True)
```

### XGBoost

```python
from xgboost import XGBRegressor


xgb = XGBRegressor(max_depth=3, learning_rate=1, n_estimators=100, random_state = 42)
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Testing MSE:', mse)
acc = xgb.score(X_test, y_test)
print('Accuracy:', acc)
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score

train, test = train_test_split(df, test_size=0.2, random_state=21)

X_train, y_train = train.copy().drop(columns=['AVG']), train['AVG']
X_test, y_test = test.copy().drop(columns=['AVG']), test['AVG']

# Create KNN classifier
knn = KNeighborsRegressor(n_neighbors = 3)
# Fit the classifier to the data
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Testing MSE:', mse)
acc = knn.score(X_test, y_test)
print('Accuracy:', acc)

#create a new KNN model
knn_cv = KNeighborsRegressor(n_neighbors=3)

#Cross validation KNN model
cv_scores = cross_val_score(knn_cv, df.loc[:, df.columns != 'AVG'], df['AVG'], cv=5)
#print each accuracy and average them
print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
```
